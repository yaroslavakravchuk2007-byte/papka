## Бонусные задачи на `networkx`

1. Зададим граф и при помощи методов `networkx` посчитаем число вершин и рёбер, радиус и диаметр главной связной компоненты, а также длины кратчайших путей между всеми парами вершин.

```python
import networkx as nx

G1 = nx.Graph()
G1.add_edges_from([
    (1, 2), (2, 3), (3, 4), (4, 1),
    (3, 5), (5, 6), (6, 7),
    (8, 9),
])

print('Число вершин:', G1.number_of_nodes())
print('Число рёбер:', G1.number_of_edges())

largest_component_nodes = max(nx.connected_components(G1), key=len)
gcc = G1.subgraph(largest_component_nodes).copy()

print('Вершины главной связной компоненты:', sorted(gcc.nodes()))
print('Радиус GCC:', nx.radius(gcc))
print('Диаметр GCC:', nx.diameter(gcc))

all_distances = dict(nx.all_pairs_shortest_path_length(G1))
print('Длины кратчайших путей между всеми парами вершин:')
for vertex in sorted(all_distances):
    print(vertex, dict(sorted(all_distances[vertex].items())))
```

2. Зададим другой граф и посчитаем его плотность, число связных компонент, словарь предшественников при обходе в глубину и все простые пути из вершины 2 в вершину 4 в полном графе из 5 вершин.

```python
import networkx as nx

G2 = nx.Graph()
G2.add_edges_from([
    (1, 2), (2, 3), (3, 4),
    (5, 6),
    (7, 8), (8, 9),
])

start_vertex = 1

print('Плотность графа:', nx.density(G2))
print('Число связных компонент:', nx.number_connected_components(G2))
print('Словарь предшественников DFS из вершины', start_vertex, ':', nx.dfs_predecessors(G2, source=start_vertex))

K5 = nx.complete_graph([1, 2, 3, 4, 5])
paths = list(nx.all_simple_paths(K5, source=2, target=4))

print('Все простые пути из 2 в 4 в полном графе K5:')
for path in paths:
    print(path)
```
